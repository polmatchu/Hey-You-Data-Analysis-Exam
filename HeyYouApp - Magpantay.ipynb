# Data Cleaning

```python
# Check for missing values
promoData.isnull().sum()
```

```python
# Drop promo_amount column (part of the instructions)
promoData = promoData.drop('promo_amount', axis=1)
```

```python
# Fill Nan values in Android, GCF, Macintosh, Tablet, Windows, X11, and iPhone with 0
# Since remaining Nan values consist of only 1.21% of the entire dataset, and only appear on 
# devices used by the customers, it is best to fill them with 0 instead
# of dropping them to not lose valuable data on the other columns.

promoData["Android"] = promoData["Android"].fillna(0)
promoData["GCF"] = promoData["GCF"].fillna(0)
promoData["Macintosh"] = promoData["Macintosh"].fillna(0)
promoData["Tablet"] = promoData["Tablet"].fillna(0)
promoData["Windows"] = promoData["Windows"].fillna(0)
promoData["X11"] = promoData["X11"].fillna(0)
promoData["iPhone"] = promoData["iPhone"].fillna(0)
```

```python
# Add day and month
promoData['trans_date'] = pd.to_datetime(promoData['trans_date'])

promoData.insert(loc=1, column='trans_month', value=promoData.trans_date.dt.month)
promoData.insert(loc=2, column='trans_day', value=(promoData.trans_date.dt.dayofweek)+1) # Monday = 1 ... Sunday = 7
```

# Exploratory Data Analysis

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

### 1. How do promos affect the other metrics? (customers, vendors, orders, amount)

```python
# Pearson Correlation
corr_data = promoData.filter(['promo_count', 'customers', 'vendors', 'orders', 'amount'])

corr = corr_data.corr()
heatmap = sns.heatmap(corr, annot=True, cmap="Oranges", fmt='.1g')
```

```python
# Regression Analysis
X = promoData['promo_count'].values.reshape(-1, 1)
y = promoData[['customers', 'vendors', 'orders', 'amount']].values
model = LinearRegression()
model.fit(promoData['promo_count'].values.reshape(-1, 1), promoData[['customers', 'vendors', 'orders', 'amount']].values)
score = model.score(X, y) # Calculate R-squared score
print("R-squared score:", score)

predictions = model.predict(X)

comparison = pd.DataFrame({'Actual_Customers': y[:, 0], 'Actual_Vendors': y[:, 1], 'Actual_Orders': y[:, 2],
                              'Actual_Amount': y[:, 3], 'Predicted_Customers': predictions[:, 0],
                              'Predicted_Vendors': predictions[:, 1], 'Predicted_Orders': predictions[:, 2],
                              'Predicted_Amount': predictions[:, 3]})
print(comparison)
```

```python
# Visualization (per day)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
plot1 = axes[0, 0] # act V pred customers
plot2 = axes[0, 1] # act V pred vendors
plot3 = axes[1, 0] # act V pred orders
plot4 = axes[1, 1] # act V pred amount

plot1.plot(promoData['trans_date'], comparison['Actual_Customers'], label='Actual Customers')
plot1.plot(promoData['trans_date'], comparison['Predicted_Customers'], label='Predicted Customers')
plot1.set_xlabel('Index')
plot1.set_ylabel('Customers')
plot1.set_title('Actual vs. Predicted Customers')
plot1.legend()

plot2.plot(promoData['trans_date'], comparison['Actual_Vendors'], label='Actual Vendors')
plot2.plot(promoData['trans_date'], comparison['Predicted_Vendors'], label='Predicted Vendors')
plot2.set_xlabel('Index')
plot2.set_ylabel('Vendors')
plot2.set_title('Actual vs. Predicted Vendors')
plot2.legend()

plot3.plot(promoData['trans_date'], comparison['Actual_Orders'], label='Actual Orders')
plot3.plot(promoData['trans_date'], comparison['Predicted_Orders'], label='Predicted Orders')
plot3.set_xlabel('Index')
plot3.set_ylabel('Orders')
plot3.set_title('Actual vs. Predicted Orders')
plot3.legend()

plot4.plot(promoData['trans_date'], comparison['Actual_Amount'], label='Actual Amount')
plot4.plot(promoData['trans_date'], comparison['Predicted_Amount'], label='Predicted Amount')
plot4.set_xlabel('Index')
plot4.set_ylabel('Amount')
plot4.set_title('Actual vs. Predicted Amount')
plot4.legend()

plt.tight_layout()
plt.show()
```

### 2. are promos helpful at all?

```python
# Comparing mean values of dependent variable with and without promos
no_promo_data = promoData[promoData['promo_count'] == 0]
no_promo_means = no_promo_data[['customers', 'vendors', 'orders', 'amount']].mean()
promo_means = promoData[promoData['promo_count'] > 0][['customers', 'vendors', 'orders', 'amount']].mean()

labels = ['Customers', 'Vendors', 'Orders', 'Amount']
x = range(len(labels))

plt.figure(figsize=(10, 6))
plt.bar(x, no_promo_means, width=0.4, align='center', label='No Promo')
plt.bar([i + 0.4 for i in x], promo_means, width=0.4, align='center', label='With Promo')
plt.xticks([i + 0.2 for i in x], labels)
plt.xlabel('Metrics')
plt.ylabel('Mean Values')
plt.title('Comparison of Metrics for Transactions With and Without Promos')
plt.legend()
plt.show()
```

```python
# Compute ANOVA to understand the impact of promos to dependent variables
import scipy.stats as stats

columns = promoData.columns[4:8]
for col in columns:
    promo_group = promoData[promoData['promo_count'] > 0][col]
    nonpromo_group = promoData[promoData['promo_count'] == 0][col]
    f_stat, p_value = stats.f_oneway(promo_group, nonpromo_group)
    print("ANOVA p-value for", col + ":", p_value)
```

### 3. What are the different customer profiles and their corresponding behavior?

```python
# Perform data exploration and analysis
fig, axes = plt.subplots(5, 1, figsize=(5, 15))
promoData[['customers', 'vendors', 'orders', 'amount', 'promo_count']].hist(ax=axes, bins=10)
plt.subplots_adjust(hspace=0.5)
```

```python
# Perform segmentation analysis
from sklearn.cluster import KMeans

X = promoData[['customers', 'vendors', 'orders', 'amount', 'promo_count']]
kmeans = KMeans(n_clusters=3) 
kmeans.fit(X)
promoData['segment'] = kmeans.labels_

# Visualize
segment_plot = promoData.plot.scatter(x='customers', y='amount', c='segment', colormap='viridis', figsize=(10, 6))
segment_plot.set_xlabel('Customers')
segment_plot.set_ylabel('Amount')
segment_plot.set_title('Customer Profiles and Transaction Amounts by Segment')
plt.show()

# Interpret
segment_counts = promoData['segment'].value_counts()
segment_means = promoData.groupby('segment')[['customers', 'vendors', 'orders', 'amount', 'promo_count']].mean()

print("Segment Counts:")
print(segment_counts)
print("\nSegment Means:")
print(segment_means)
```

```python
# Devices Distribution
device_data = promoData[['Android', 'GCF', 'Macintosh', 'Tablet', 'Windows', 'X11', 'iPhone']]
device_counts = device_data.sum()
total_devices = device_counts.sum()
device_percentages = (device_counts / total_devices) * 100

device_plot = device_percentages.plot(kind='bar', figsize=(10, 6))
device_plot.set_title("Customer Devices Distribution", fontsize=19)
device_plot.set_xlabel("Device Type", fontsize=17)
device_plot.set_ylabel("Percentage", fontsize=17)

for p in device_plot.patches:
    height = p.get_height()
    device_plot.text(p.get_x() + p.get_width() / 2., height + 0.5, '{:1.2f}%'.format(height),
                     ha="center", fontsize=14)

plt.ylim(0, 100)
plt.show()
```

```python
# Blacklisted Distribution
blkList_data = promoData[['Blacklisted', 'Not Blacklisted']]
blkList_counts = blkList_data.sum()
total_blkList = blkList_counts.sum()
blkList_percentages = (blkList_counts / total_blkList) * 100

blkList_plot = blkList_percentages.plot(kind='bar', figsize=(10, 6))
blkList_plot.set_title("Customer Blacklist Distribution", fontsize=19)
blkList_plot.set_ylabel("Percentage", fontsize=17)

for p in blkList_plot.patches:
    height = p.get_height()
    blkList_plot.text(p.get_x() + p.get_width() / 2., height + 0.5, '{:1.2f}%'.format(height),
                     ha="center", fontsize=14)

plt.ylim(0, 100)
plt.show()
```

### 4. Additional Findings

```python
# Temporal Analysis
promoData['trans_date'] = pd.to_datetime(promoData['trans_date'], format='%d/%m/%Y')
daily_customers = promoData.groupby('trans_date')['customers'].sum()
plt.figure(figsize=(20, 5))
plt.plot(daily_customers, marker='+', c="darkorange")
plt.plot(daily_customers.rolling(window=30, center=True).mean(), c="red")
plt.xticks(rotation=90)
plt.title("Total Number of Customers per Day")
plt.xlabel("Date")
plt.ylabel("Number of Customers")
plt.legend(['Daily Customers', '30-Day Rolling Average'])
plt.show()
```

```python
# Top Days with the Least Customers
grouped_customers = promoData.groupby('trans_date')['customers'].sum()
top_10_days_least_customers = grouped_customers.nsmallest(10)
print("Top 10 Days with the Least Number of Customers:")
for date, customers in top_10_days_least_customers.iteritems():
    formatted_date = date.strftime('%Y-%m-%d')
    print(f"{formatted_date}: {customers} customers")
```

```python
# Check the business days and hours
grouped_df = promoData.groupby(['trans_day', 'trans_hour'])['customers'].mean().reset_index()
grouped_df = grouped_df.pivot('trans_hour', 'trans_day', 'customers')
plt.figure(figsize=(10, 6))
sns.heatmap(grouped_df, cmap='YlGnBu', annot=True, fmt='.2f')
plt.title('Mean Number of Customers per Hour and Day', fontsize=15)
plt.xlabel('Day of Week')
plt.ylabel('Hour of Day')
plt.show()
```
